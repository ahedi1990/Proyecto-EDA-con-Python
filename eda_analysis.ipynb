```python
import pandas as pd
import seaborn as sns
```

```python
# Cargar CSV
df_bank = pd.read_csv('C:/Users/Usuario/Desktop/MIO/CURSO/marketing-analysis-project/DATOS/Crudo/bank-additional.csv', sep=',' )

# Cargar Excel con múltiples hojas
customer_details = pd.ExcelFile('C:/Users/Usuario/Desktop/MIO/CURSO/marketing-analysis-project/DATOS/Crudo/customer-details.xlsx')
sheets = pd.read_excel(customer_details, sheet_name=None)  # Carga todas las hojas como diccionario

# Unimos las tres hojas en un solo DataFrame
df_customers = pd.concat(sheets.values(), ignore_index=True)
```

```python
# ===================
# LIMPIEZA df_bank
# ===================

# Revisar valores nulos
print("Valores nulos en df_bank:")
print(df_bank.isnull().sum())

# Convertir 'y', 'default', 'housing', 'loan' a booleano
bin_cols = ['y', 'default', 'housing', 'loan']
for col in bin_cols:
    df_bank[col] = df_bank[col].replace({'yes': 1, 'no': 0, 'unknown': pd.NA})

# Convertir 'date' a formato datetime
df_bank['date'] = pd.to_datetime(df_bank['date'], errors='coerce')

# Eliminar duplicados por ID si hay
df_bank = df_bank.drop_duplicates(subset='id_')

# Comprobamos el tipo de datos
print("\nTipos de datos en df_bank:")
print(df_bank.dtypes)
```

```python
# ===================
# LIMPIEZA df_customers
# ===================

# Convertir fecha de cliente a datetime
df_customers['Dt_Customer'] = pd.to_datetime(df_customers['Dt_Customer'], errors='coerce')

# Renombrar columna ID para que coincida con df_bank
df_customers = df_customers.rename(columns={'ID': 'id_'})

# Eliminar duplicados por ID si hay
df_customers = df_customers.drop_duplicates(subset='id_')

# Revisar valores nulos
print("\nValores nulos en df_customers:")
print(df_customers.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo general
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# ====================
# UNIÓN DE DATOS
# ====================
# Combinar ambos datasets usando 'id_'
df_full = pd.merge(df_bank, df_customers, on='id_', how='left')

# ====================
# ANÁLISIS DESCRIPTIVO
# ====================
print("\n=== Estadísticas descriptivas (numéricas) ===")
print(df_full.describe())

print("\n=== Distribución de variables categóricas ===")
cat_cols = ['job', 'marital', 'education', 'contact', 'poutcome']
for col in cat_cols:
    print(f"\n{col}:\n{df_full[col].value_counts()}")

# ====================
# VISUALIZACIONES
# ====================

# Histograma de edad
sns.histplot(df_full['age'], kde=True, bins=20)
plt.title("Distribución de Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

# Boxplot de ingresos por si suscribió o no
sns.boxplot(data=df_full, x='y', y='Income')
plt.title("Ingreso Anual según Suscripción")
plt.xlabel("Suscripción")
plt.ylabel("Ingreso")
plt.show()

# Conteo por ocupación
plt.figure(figsize=(12,5))
sns.countplot(data=df_full, x='job', order=df_full['job'].value_counts().index)
plt.title("Distribución por Ocupación")
plt.xticks(rotation=45)
plt.show()

# Matriz de correlación numérica
plt.figure(figsize=(10,8))
sns.heatmap(df_full.select_dtypes(include='number').corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matriz de Correlación")
plt.show()

# Distribución de clientes que contrataron el producto
sns.countplot(data=df_full, x='y')
plt.title("¿Cliente suscribió el producto?")
plt.xlabel("Suscripción")
plt.ylabel("Cantidad")
plt.show()
```
